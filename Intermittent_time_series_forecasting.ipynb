### **1.3 Traditional time series vs Intermittent time series forecasting**


---


#### *Time Series Forecasting:*

Time series forecasting is a general methodology used to predict future values of a variable based on its historical values, which are collected at equally spaced time intervals. This approach assumes that there is a certain degree of continuity and consistency in the patterns of the data over time. Traditional time series methods like ARIMA, exponential smoothing, and state-space models excel at capturing trends, seasonality, and cyclic behavior within the data.

**Characteristics:**

* Assumes consistent and regular patterns.
* Designed to handle continuous and smooth data variations.
* Effective for data with clear trends, seasonality, and cyclic fluctuations.

**Applications:**

* Stock price prediction.
* Temperature forecasting.
* Sales prediction for regularly consumed products.

####*Intermittent Time Series Forecasting:*

Intermittent time series forecasting, on the other hand, deals with data that exhibits sporadic or irregular patterns. This type of data, often referred to as "lumpy data," features periods of inactivity interspersed with sudden bursts of activity. These sporadic patterns can be influenced by factors like seasonality, demand volatility, and supply chain disruptions. Traditional time series forecasting methods struggle to accurately capture and predict these intermittent patterns.

**Characteristics:**

* Involves sporadic or infrequent observations.
* Irregular activity periods followed by periods of inactivity.
* Can be influenced by factors that lead to abrupt changes in demand or activity.

**Applications:**

* Demand forecasting for spare parts.
* Sales prediction for products with irregular buying patterns.
* Resource allocation for emergency services during sporadic events.

####*Key Distinctions:*

* **Data Patterns:** The most significant distinction lies in the nature of the data patterns. Time series forecasting assumes continuous and consistent patterns, while **intermittent forecasting deals with irregular, sporadic, and intermittent data patterns.**

* **Methodology:** Traditional time series methods are designed to model and predict continuous trends and fluctuations. Intermittent forecasting methods are tailored to handle the unique challenges posed by sporadic and intermittent patterns.

* **Challenges:** Intermittent forecasting faces challenges related to capturing abrupt changes in activity levels and dealing with periods of inactivity, which are not as prominent in regular time series data.

* **Techniques:** Intermittent forecasting often requires specialized techniques that can account for the sporadic nature of the data, such as Croston's method, aggregate-disaggregate approaches, and probabilistic models.

In summary, intermittent time series forecasting is a specialized subset of time series forecasting that addresses the complexities of sporadic and irregular data patterns. While traditional time series methods excel in handling continuous trends, they struggle to accurately capture the nuances of data characterized by intermittent activity.

#### **1.4.1  Croston's Method**

In 1972, J.D. Croston published “Forecasting and Stock Control for Intermittent Demands,” an article introducing a new technique to forecast products with intermittent demand. His idea could be summarized in three simple steps:
- Evaluate the average demand level when there is a demand occurrence.
- Evaluate the average time between two demand occurrences.
- Forecast the demand as the demand level (when there is an occurrence) multiplied by the probability of having an occurrence.

Let’s do this step-by-step with an example:

-  If we estimate that we have a demand occurrence every four periods on average (i.e., a 25% chance to have an occurrence),
- and we expect that —when there is a demand occurrence — the average demand level is five units,
- then, we forecast the demand to be 5 x 0.25 = 1.25 per period going forward.

Croston’s method is one of the most common approaches to forecasting spare time series. It often acts as a baseline model to evaluate more complex methods.

With Croston’s method, two series are constructed from the original series:

* A series containing the time periods with only zero values
- A series containing time periods with non-zero values

This disaggregation approach helps in capturing the intermittent patterns more accurately.

Let's take an example of weekly sales dataset

- The first step is to use the Simple Exponential Smoothing (SES) method to predict the values for the days when there is demand.

- The SES is fit to a series of values that excludes the zeros, then it forecasts the next value in the series (the value we should see the next day with sales).

- The second step is to use a different SES model to predict the intervals between days with sales.

- This step is important because it helps us to predict the number of days until the next sale occurs.

- We forecast one-step ahead, so we will get the number of days until the next sale occurs.

- The final step is to divide the first forecast (sales value) by the second forecast (number of days until the next sale).

- This gives us an estimate of the demand for each day.

- If the second forecast is zero, we use the first forecast as our final prediction.

- This is because, if the second forecast is zero, it means that there will be demand every day and our final prediction will be the same as the first forecast.

#### **1.4.2 Aggregate-Disaggreagate Intermittent Demand Approach (ADIDA)**

The Aggregate-Disaggregate Intermittent Demand Approach, or ADIDA, aims to remove intermittence by aggregating the series at a lower frequency.

For example, if hourly data has zero values, then summing over 24 hours to get daily data might get rid of the zero values. The same logic applies to intermittent daily data that we could aggregate to weekly data to remove periods with zero values.

```python
agg_sim_df = sim_df['y'].rolling(window=5, step=5).sum()
agg_sim_df.reset_index(drop=True)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,8))

ax1.bar(sim_df.index, sim_df['y'], color='lightgray')
ax1.set_ylabel('Value')
ax1.set_xlabel('Time steps')

ax2.bar(agg_sim_df[1:].index, agg_sim_df[1:])
ax2.set_ylabel("Aggregated value")
ax2.set_xlabel('Buckets')

plt.tight_layout()
```

In the figure above, we can see the effect of aggregation on our simulated data. Here, we aggregate over five time steps. The resulting aggregated series, shown in the bottom plot, is not intermittent anymore, since we got rid of all zero values.

Once the data is aggregated, simple exponential smoothing is again used to forecast the aggregated series.

Then, we disaggregate the predictions to bring them back to the original frequency. For example, if hourly data was aggregated to daily data, then each prediction would be divided by 24 (since there are 24 hours in a day) to get the disaggregated predictions.

While there is no clear answer on how to choose the aggregation level, one way that is implemented in statsforecast is to calculate the the length of all intervals between non-zero values, and take the average of the intervals as the aggregate level.

For example, if your intermittent series has intervals between non-zero values of [3, 5, 4], then the aggregate level would be 4.

In the best case scenario, this method completely removes the intermittency. Otherwise, only a few zero values will remain, which should not greatly impact exponential smoothing

```python
from statsforecast.models import ADIDA
models = [CrostonOptimized(), ADIDA()]

sf = StatsForecast(
    df=sim_df,
    models=models,
    freq='H',
    n_jobs=-1
)

cv_df = sf.cross_validation(
    df=sim_df,
    h=1,
    step_size=1,
    n_windows=50
)

cv_df.index = np.arange(50, 100, 1)
```

Now, let’s implement ADIDA on our simulated data and see the predictions we obtain.

Using statsforecast, the implementation remains straightforward, as we simply need to change the model, but the pipeline stays the same.

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.bar(sim_df.index, sim_df['y'], color='lightgray')
ax.plot(cv_df.index, cv_df['CrostonOptimized'], ls='--', label='Croston (optimized)')
ax.plot(cv_df.index, cv_df['ADIDA'], ls=':', label='ADIDA')
ax.set_ylabel('Value')
ax.set_xlabel('Time steps')
ax.legend(loc='best')
plt.xlim(40, 100)

plt.tight_layout()
```

In the figure above, we can see that ADIDA reacts much more to periods with zero values. While the predictions from Croston will remain constant if zero values are observed, ADIDA will gradually decrease the prediction curve, and so it approaches more the actual data.

#### **1.4.3 Teunter-Syntetos-Babais(TSB)**

The Teunter-Syntetos-Babais model, or TSB, proposes an improvement over Croston’s method.

As we have seen earlier, predictions from Croston’s method stay constant during period with zero values. This means that the predictions can become outdated with many periods of zero values.

In other words, Croston’s method ignores the risk of obsolescence, which occurs when non-zero values are separated by longer and longer zero demand periods.

This is especially important in inventory management of low-demand products, as companies can hold stocks of unused inventory for many years, which comes at a cost. They must therefore assess the risk of obsolescence to determine if they can get rid of dead stock or not.

This is where the TSB model comes in. Instead of considering the demand interval, which are the periods of zero values, it will consider the demand probability

```python
from statsforecast.models import TSB

models = [TSB(0.1, 0.1), CrostonClassic()]

sf = StatsForecast(
    df=sim_df,
    models=models,
    freq='H',
    n_jobs=-1
)

cv_df = sf.cross_validation(
    df=sim_df,
    h=1,
    step_size=1,
    n_windows=50
)

cv_df.index = np.arange(50, 100, 1)
```

While this seems to be a small difference, it can actually have a big impact. With Croston’s method, the demand interval can only be updated once we observe a non-zero value.

On the other hand, the demand probability is updated at every time step, making the model more flexible.

To make a prediction, the model also creates two series from the original series:

- One containing only non-zero values (also called demand)
- The other is for the demand probability

Predictions for each series is done using simple exponential smoothing. Then, the final prediction is obtained by multiplying the demand by the demand probability.

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.bar(sim_df.index, sim_df['y'], color='lightgray')
ax.plot(cv_df.index, cv_df['TSB'], ls='--', label='TSB')
ax.plot(cv_df.index, cv_df['CrostonClassic'], ls=':', label='Croston')
ax.set_ylabel('Value')
ax.set_xlabel('Time steps')
ax.legend(loc='best')
plt.xlim(40, 100)

plt.tight_layout()
```

Looking at the figure above, we can see how using the demand probability makes a big difference, as the predictions are decreasing during no-demand periods, instead of being constant.

#### **1.5 LSTM for intermittent time series**


---

In the realm of time series forecasting, the Long Short-Term Memory (LSTM) network has emerged as a powerful and versatile tool. LSTM is a type of recurrent neural network (RNN) architecture specifically designed to handle sequential data, making it well-suited for forecasting tasks involving complex patterns and dependencies.

**Introduction to RNNs:**

Recurrent Neural Networks (RNNs) are a class of neural networks designed to process sequences of data by maintaining a hidden state that captures information from previous time steps. RNNs are capable of modeling temporal dependencies, making them suitable for tasks involving sequential data like time series.

Vanilla RNNs suffer from the vanishing gradient problem, where information from distant past time steps tends to be lost during training. This hinders their ability to capture long-range dependencies.

**Introduction to LSTM:**

LSTM was introduced to overcome the limitations of vanilla RNNs. It incorporates memory cells and gating mechanisms that enable it to capture long-range dependencies while avoiding the vanishing gradient problem.

**Key Components of LSTM:**

- Cell State: The cell state acts as the memory of the LSTM. It can store and carry information across different time steps.
- Forget Gate: Determines what information to discard from the cell state, addressing the vanishing gradient problem.
- Input Gate: Updates the cell state with new information.
- Output Gate: Determines the output based on the current cell state.

**Handling Intermittent Time Series Data:**

LSTM's ability to capture long-term dependencies and handle sequences of varying lengths makes it suitable for forecasting intermittent time series data. It can learn to identify patterns even when periods of low or zero activity are present.

```python

def create_dataset(dataset, window_size = 1):
    data_X, data_Y = [], []
    for i in range(len(dataset) - window_size - 1):
        a = dataset[i:(i + window_size), 0]
        data_X.append(a)
        data_Y.append(dataset[i + window_size, 0])
    return(np.array(data_X), np.array(data_Y))
```

This code defines a function create_dataset that prepares a time series dataset for training a sequence-based forecasting model. The function takes the input dataset and a window size as arguments. Here's how the function works:

- Inputs:

dataset: The normalized time series dataset you've previously prepared.
window_size: An integer representing the size of the input window for creating sequences. It determines how many previous time steps the model will consider when making a prediction.

- Output:

The function returns two arrays, data_X and data_Y, which will be used for training the forecasting model. data_X contains the input sequences, and data_Y contains the corresponding target values.

- Loop to Create Sequences:

The function uses a loop to iterate over the range of indices from 0 to (len(dataset) - window_size - 1).

For each iteration, it extracts a sequence of length window_size from the original dataset. This sequence is represented by the variable a.
The data_X list is populated with these sequences (a), which will be used as input features for the model.


- Target Values:

For each iteration, the target value (the next value in the sequence after the window) is extracted from the dataset and added to the data_Y list. This will be the value the model needs to predict.


- Output Conversion:

After the loop, the function converts the data_X and data_Y lists into numpy arrays using np.array().
The function then returns the two arrays as a tuple: (data_X, data_Y).
The purpose of this function is to transform the original time series data into sequences that can be fed into a recurrent neural network, such as an LSTM model. The sequences of input values (data_X) will be used to predict the target values (data_Y) using the forecasting model.

 The window_size parameter determines how many time steps the model looks back in the past to make predictions.

```python
# Create test and training sets for one-step-ahead regression.
window_size = 1
train_X, train_Y = create_dataset(train, window_size)
test_X, test_Y = create_dataset(test, window_size)
print("Original training data shape:")
print(train_X.shape)

# Reshape the input data into appropriate form for Keras.
train_X = np.reshape(train_X, (train_X.shape[0], 1, train_X.shape[1]))
test_X = np.reshape(test_X, (test_X.shape[0], 1, test_X.shape[1]))
print("New training data shape:")
print(train_X.shape)
```

```python
def fit_model(train_X, train_Y, window_size = 1):
    model = Sequential()

    model.add(LSTM(2000,activation = 'tanh', input_shape = (1, window_size)))
    model.add(Dropout(0.2))
    model.add(Dense(500))
    model.add(Dropout(0.4))
    model.add(Dense(500))
    model.add(Dropout(0.4))
    model.add(Dense(400))
    model.add(Dropout(0.4))
    model.add(Dense(1, activation = 'linear'))
    model.compile(loss = "mean_squared_error",
                  optimizer = "adam")
    model.fit(train_X,
              train_Y,
              epochs = 10,
              batch_size = 64,
              )

    return(model)

# Fit the first model.
model1 = fit_model(train_X, train_Y, window_size)
```

A Dropout layer with a dropout rate of 0.2 to prevent overfitting.
Several Dense layers with different numbers of units (500, 500, 400) and Dropout layers with dropout rates of 0.4.

A final Dense layer with a single unit and a linear activation function for regression.

The model is compiled with the "mean_squared_error" loss function and the "adam" optimizer.

#### **1.6 DeepAR**


Autoregressive Modeling:

Autoregressive modeling is a fundamental concept in time series analysis, and it plays a crucial role in the functioning of the DeepAR algorithm. In autoregressive models, the current value of a time series is predicted based on its own previous values. The idea is that past observations contain information about the future behavior of the series.

1.6.2 AutoRegressive Integrated Moving Average (ARIMA):

A classic example of an autoregressive model is the AutoRegressive Integrated Moving Average (ARIMA) model. ARIMA combines autoregression (AR), differencing (I), and moving average (MA) components to model and forecast time series data. It captures trends, seasonality, and autocorrelation within the data.

1.6.3 Autoregressive Nature of DeepAR:

DeepAR's autoregressive nature aligns with its autoregressive modeling principle. It learns to generate forecasts by considering the historical data leading up to the point of prediction. This allows DeepAR to capture dependencies and patterns in the data, making it particularly effective for forecasting tasks where the past behavior significantly influences future outcomes.

1.6.4 Handling Intermittent Time Series with DeepAR:

Intermittent time series data is characterized by periods of inactivity or low activity interspersed with occasional spikes or bursts. DeepAR is well-suited for such data due to its flexibility in capturing irregular patterns and its probabilistic forecasting approach.

Here's how DeepAR is used for intermittent time series forecasting:

- Sequence Generation: DeepAR prepares the data by creating sequences that capture the intermittent behavior. Each sequence represents a portion of the time series and includes a mix of both active and inactive periods.

- Encoding Features: DeepAR can incorporate additional features such as categorical variables or dynamic covariates related to the time series. This enables the model to capture contextual information and enhance predictions.

- Probabilistic Forecasts: DeepAR produces probabilistic forecasts, which align well with the uncertainty inherent in intermittent time series data. Instead of providing a single point estimate, the model generates a distribution of possible future values.

- Quantile Loss Minimization: In intermittent time series, accurate forecasting of extreme values (e.g., spikes) is crucial. DeepAR can be trained to minimize quantile loss, ensuring that the model assigns higher probabilities to extreme quantiles where intermittent spikes occur.

- Uncertainty Estimation: DeepAR estimates the uncertainty associated with its forecasts. This is particularly valuable in intermittent time series forecasting, where the irregular nature of events can lead to high uncertainty in predictions.

- Practical Applications: DeepAR's ability to handle intermittent time series makes it applicable in scenarios like retail demand forecasting (e.g., predicting sudden spikes in sales during promotions), energy consumption prediction (e.g., forecasting sudden surges in usage), and fraud detection (e.g., identifying irregular patterns in transaction data).

DeepAR's architecture is a key factor in its success as a forecasting technique. It's designed to capture the intricacies of time series data, handle uncertainties, and produce accurate probabilistic forecasts. Let's dive deeper into the layers that make up DeepAR's architecture:

1.6.5 Encoder-Decoder Framework:

DeepAR follows an encoder-decoder architecture, which is a common structure in sequence-to-sequence modeling. This framework consists of two main components: the encoder and the decoder.

- Encoder: The encoder processes the historical input sequence, extracting relevant patterns and features. This process involves learning a compact representation of the input data that captures its temporal dependencies.

- Decoder: The decoder generates future predictions based on the encoded information from the encoder. It's responsible for producing the forecasted values and their associated probability distributions.

1.6.6 Recurrent Neural Network (RNN) Layers:

The core of DeepAR's architecture involves the use of recurrent neural network (RNN) layers, which are well-suited for sequence data analysis due to their ability to capture temporal dependencies.

- LSTM (Long Short-Term Memory): DeepAR employs Long Short-Term Memory (LSTM) units as the building blocks of its RNN layers. LSTMs excel at capturing long-range dependencies in sequential data while addressing the vanishing gradient problem.

Multi-Step Prediction:
DeepAR is designed for multi-step prediction, meaning it generates forecasts for multiple future time steps. This is achieved by sequentially feeding the model's own predictions back as input for generating subsequent predictions.

1.6.7 Probabilistic Forecasting:
One of DeepAR's defining features is its ability to produce probabilistic forecasts. Instead of providing a single point estimate, the model generates a probability distribution for each future time step. This distribution encapsulates the range of possible outcomes, including the uncertainty associated with the predictions.

- Quantile Regression Loss: To generate these probabilistic forecasts, DeepAR minimizes a quantile regression loss function during training. This loss encourages the model to produce predictions that match specific quantiles of the predictive distribution.

1.6.8 Contextual Covariates and Features:
DeepAR can incorporate contextual covariates and additional features to enhance forecasting accuracy. These features can include dynamic variables that influence the time series behavior or categorical variables that capture context.

1.6.9 Embeddings and Attention Mechanisms:
To handle categorical variables, DeepAR employs embeddings and attention mechanisms. Embeddings transform categorical inputs into continuous vector representations, while attention mechanisms help the model focus on relevant inputs when making predictions.

1.6.10 Hyperparameter Tuning and Regularization:
DeepAR's architecture also includes hyperparameters that control the model's behavior, such as the number of LSTM units, the number of layers, learning rates, and dropout rates. Hyperparameter tuning is essential for achieving optimal performance.

1.6.11 Practical Use and Interpretability:
While the architecture of DeepAR is complex, it offers robust and accurate forecasting capabilities. However, the inherent complexity of deep learning models can make them less interpretable compared to traditional statistical methods. Efforts are being made to develop techniques that improve the interpretability of deep learning models, including DeepAR.

Conclusion

DeepAR's architecture showcases its versatility and sophistication in handling time series forecasting challenges. Its combination of an encoder-decoder framework, LSTM units, probabilistic forecasting, contextual covariates, and attention mechanisms empowers DeepAR to capture dependencies, generate accurate forecasts, and provide valuable insights into uncertainty. Understanding this architecture not only enhances our appreciation of DeepAR's capabilities but also equips us with the knowledge needed to leverage its power for accurate and informed decision-making in a variety of industries.
